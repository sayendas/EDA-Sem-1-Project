**Analysis of covid_19_india csv file**

```python
c19i = pd.read_csv(io.BytesIO(uploaded['covid_19_india.csv']), sep=',')
```

```python
c19i
```

```python
c19i.dtypes
```

```python
c19i.isnull().sum()
```

```python
c19i['State/UnionTerritory'].unique()   #we have some weird values in this column which we need to clean
```

```python
c19i[c19i['State/UnionTerritory']=='Maharashtra']
```

```python
c19i[c19i['State/UnionTerritory']=='Maharashtra***']
```

```python
#we need to replace values like Bihar****,Madhya Pradesh***,Maharashtra*** with there actual state names by removing the *s

c19i['State/UnionTerritory'].replace("Bihar****","Bihar",inplace=True)
c19i['State/UnionTerritory'].replace("Madhya Pradesh***","Madhya Pradesh",inplace=True)
c19i['State/UnionTerritory'].replace("Maharashtra***","Maharashtra",inplace=True)
```

```python
c19i['State/UnionTerritory'].unique()    #we have now removed the weird values from the State column,we can see there are many spelling mistakes in the State column
```

```python
#we have seperate rows for [Dadra and Nagar Haveli] and [Daman & Diu], we need to remove those rows as as both of these are cosidered as a single union territory
#for which we already have seperate rows

droprows=c19i[(c19i['State/UnionTerritory']=='Dadra and Nagar Haveli') | (c19i['State/UnionTerritory']=='Daman & Diu')].index

c19i.drop(droprows,inplace=True)

print(c19i['State/UnionTerritory'].unique())     #we now see that there are no seperate rows for [Dadra and Nagar Haveli] and [Daman & Diu]
```

```python
#we also see spelling mistakes in State column, such as Telengana which should be Telangana etc

c19i['State/UnionTerritory'].replace("Telengana","Telangana",inplace=True)
c19i['State/UnionTerritory'].replace("Karanataka","Karnataka",inplace=True)
c19i['State/UnionTerritory'].replace("Himanchal Pradesh","Himachal Pradesh",inplace=True)
```

```python
#in the file, the number of cured,deaths and confirmed cases are cumulative to date, so for analysis of total numbers, we only have to consider the last date value

plt.figure(figsize=[20,10])

ax=sns.barplot(data=c19i[c19i['Date']=='2021-08-11'].sort_values('Cured',ascending=False),x='Cured',y='State/UnionTerritory',errorbar=None)
ax.bar_label(ax.containers[0],fmt='%i')

plt.title('Cured people per State')
plt.xscale('log')
plt.show()
```

Maharastra has the most number of cured patients,followed by Kerala and then Karnataka

```python
plt.figure(figsize=[20,10])

ax=sns.barplot(data=c19i[c19i['Date']=='2021-08-11'].sort_values('Deaths',ascending=False),x='Deaths',y='State/UnionTerritory',errorbar=None)
ax.bar_label(ax.containers[0],fmt='%i')

plt.title('Number of expired people per State')
plt.xscale('log')
plt.show()
```

Maharastra has the highest death rate in India, followed by Karnataka and then Tamil Nadu

```python
plt.figure(figsize=[20,10])

ax=sns.barplot(data=c19i[c19i['Date']=='2021-08-11'].sort_values('Confirmed',ascending=False),x='Confirmed',y='State/UnionTerritory',errorbar=None)
ax.bar_label(ax.containers[0],fmt='%i')

plt.title('Confirmed cases per State')
plt.xscale('log')
plt.show()
```

Maharastra also has the maximum number of covid confirmed cases,followed by Kerela and then Karnataka

```python
#As Maharashtra, Karnataka and Tamil Nadu are 3 states with very high deaths, we can analyse over a period how covid is affecting deaths in these 3 states
#we have a Date column from which we can extract month and year information which can be used for analysis

c19i['Year'] = c19i['Date'].str[0:4]

c19i['Month'] = c19i['Date'].str[5:7]

c19i['Month and Year'] = c19i['Month'] +' '+c19i['Year']
```

```python
c19i.columns
```

```python
c19i_top3deaths=c19i[(c19i['State/UnionTerritory']=='Karnataka') | (c19i['State/UnionTerritory']=='Maharashtra') | (c19i['State/UnionTerritory']=='Tamil Nadu')]

plt.figure(figsize=[20,10])

sns.lineplot(data=c19i_top3deaths,x='Month and Year',y='Deaths',err_style=None,hue='State/UnionTerritory',style='State/UnionTerritory',marker=True)

plt.show()
```

At the start of March 2020, all 3 states had almost no deaths.

For Maharastra, covid deaths started happening from April 2020 and has kept increasing since

For Karnataka and Tamil Nadu, deaths started happening from June 2020

Tamil Nadu has witnessed a higher death count compared to Karnataka from June 2020 to October 2020, however after that period from November 2020 to April 2021, the death count has remained the same for the 2 States

After April 2021, there is increase in death count for Karnataka as compared to Tamil Nadu

```python
#as we have seen some insights utilising a single dataset, we can now join 2 datasets to get insights

print(c19i)

print(pic)
```

```python
#we can find the state having highest ratio of confirmed people as compared to total population

c19i_pic_inner=pd.merge(c19i,pic,left_on='State/UnionTerritory',right_on='State / Union Territory',how='inner')

c19i_pic_inner['confirmed_people_ratio']=c19i_pic_inner['Confirmed']/c19i_pic_inner['Population']    #add new column to check ratio of confirmed covid cases


#as already seen,we have to take date as 2021-08-11 in the plot as data in the c19i dataset is incremental and we have confirmed cases upto 2021-08-11 in dataset
#for each state

plt.figure(figsize=[20,10])

ax=sns.barplot(data=c19i_pic_inner[c19i_pic_inner['Date']=='2021-08-11'].sort_values('confirmed_people_ratio',ascending=False),x='confirmed_people_ratio',y='State/UnionTerritory',errorbar=None)
ax.bar_label(ax.containers[0])

plt.title('Ratio of confirmed covid cases per state')
plt.xlabel('Ratio of confirmed covid cases')

plt.show()
```

We can see that Lakshwadeep, Goa and Kerala are the top 3 places where covid confirmed cases were higher compared to their population

```python
#similar to ratio of confirmed cases, we can also get insights for ratio of deaths and ratio of cured people

c19i_pic_inner['dead_people_ratio']=c19i_pic_inner['Deaths']/c19i_pic_inner['Population']

c19i_pic_inner['cured_people_ratio']=c19i_pic_inner['Cured']/c19i_pic_inner['Population']


plt.figure(figsize=[20,10])

ax=sns.barplot(data=c19i_pic_inner[c19i_pic_inner['Date']=='2021-08-11'].sort_values('dead_people_ratio',ascending=False),x='dead_people_ratio',y='State/UnionTerritory',errorbar=None)
ax.bar_label(ax.containers[0])

plt.title('Ratio of covid death cases per state')
plt.xlabel('Ratio of covid death cases')

plt.show()
```

We can see that Goa, Delhi and Puducherry have are top 3 places with high ratio of death cases

```python
plt.figure(figsize=[20,10])

ax=sns.barplot(data=c19i_pic_inner[c19i_pic_inner['Date']=='2021-08-11'].sort_values('cured_people_ratio',ascending=False),x='cured_people_ratio',y='State/UnionTerritory',errorbar=None)
ax.bar_label(ax.containers[0])

plt.title('Ratio of covid cured cases per state')
plt.xlabel('Ratio of covid cured cases')

plt.show()
```

We can see that Lakshwadeep, Goa and Kerala are the top 3 places where covid cured cases were higher compared to their population

```python
#we can also see the which states are most effective in administering vaccine doses as we have the cvs and pic dataset which contain the total doses administered
#and total population of each state


cvs_pic_inner=pd.merge(cvs,pic,left_on='State',right_on='State / Union Territory',how='inner')

cvs_pic_inner['total_doses_total_population']=cvs_pic_inner['Total Doses Administered']/cvs_pic_inner['Population']


plt.figure(figsize=[20,10])

ax=sns.barplot(data=cvs_pic_inner[cvs_pic_inner['Updated On']=='09/08/2021'].sort_values('total_doses_total_population',ascending=False),x='total_doses_total_population',y='State / Union Territory',errorbar=None)
ax.bar_label(ax.containers[0])

plt.title('Ratio of total doses administered for total population per state')
plt.xlabel('Ratio of total doses administered for total population')

plt.show()
```

We can see that Dadar and Nagar Havali and Daman and Diu, Sikkim and Lakhshwadeep are the top 3 places which are most effective in administering vaccine doses. These are the only 3 places where the ratio is greater than 1 which means 100% population has taken at least a single covid dose.

```python
# we can get the ratio of number of total beds(rural+urban+total public beds) to confirmed cases of each state to get how well each state has its hospital management

#print(hbi)

#print(c19i)

#total number of total beds(urban beds + rural beds + total public beds)

hbi['Total_number_of_beds'] = hbi['NumPublicBeds_HMIS']+hbi['NumRuralBeds_NHP18']+hbi['NumUrbanBeds_NHP18']

hbi_c19i_inner=pd.merge(hbi,c19i,left_on='State/UT',right_on='State/UnionTerritory',how='inner')
```

```python
hbi_c19i_inner
```

```python
hbi_c19i_inner['ratio of total hospital beds to confirmed covid cases'] = hbi_c19i_inner['Total_number_of_beds']/hbi_c19i_inner['Confirmed']
```

```python
plt.figure(figsize=[20,10])

ax=sns.barplot(data=hbi_c19i_inner[hbi_c19i_inner['Date']=='2021-08-11'].sort_values('ratio of total hospital beds to confirmed covid cases',ascending=False),x='ratio of total hospital beds to confirmed covid cases',y='State/UnionTerritory',errorbar=None)
ax.bar_label(ax.containers[0])

plt.title('Ratio of total hospital beds to confirmed covid cases for each state')
plt.xlabel('Ratio of total hospital beds to confirmed covid cases')

plt.show()
```

We can see Nagaland, Meghalaya and Tripura are the 3 states which have the best hospital management as they have high ratio of total number of beds to number of confirmed covid cases

**Inferences from above data**

1) Goa, Delhi and Puducherry are the 3 places which are worst affected as these 3 states/UT have the highest death ratio compared to their total population.

2) Dadra and Nagar Havali and Daman and Diu, Sikkim and Lakhswadeep are the 3 States/UT which are most effective in adminstering the covid doses.
